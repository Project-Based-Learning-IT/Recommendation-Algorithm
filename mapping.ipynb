```python
import pandas as pd
```

```python
df = pd.read_csv('mapping.csv',header=None)
df.columns = ['Domain','Skills']
df.head()
```

```python
mappingDict = {}
Domains = df['Domain'].values
Skills = df['Skills'].values
# for domain in Domains:
# print(Skills)
# print(Domains)
for i in range(len(Domains)):
    mappingDict[Domains[i]] = list(map(lambda x:x.strip(), Skills[i].split(',')))
    # print(list(map(lambda x:x.strip(), Skills[i].split(','))))
```

```python
file2 = open('names.txt','r')
namesList = file2.read()
# print(list(map(lambda x:x.strip(),namesList.strip('][').replace("'","").split(','))))
names = list(map(lambda x:x.strip(),namesList.strip('][').replace("'","").split(',')))
```

```python
print(f'need to make {len(names)} user profiles')
```

```python
import random
```

```python
AllSkills = set()
for skills in mapDict.values():
    for skill in skills:
        AllSkills.add(skill.lower())
AllSkills = list(AllSkills)
print(f'Total skills - {len(AllSkills)}')
```

```python
minSkillCount = 0
maxSkillCount = len(AllSkills)
minToSelect = 5
maxToSelect = 16
data = {}
for user in names:
    skillCntForThisUser = random.randint(minToSelect,maxToSelect)
    skillIndx = random.sample(range(minSkillCount,maxSkillCount),skillCntForThisUser)
    skillNames = []
    for index in skillIndx:
        skillNames.append(AllSkills[index])
    data[user] = skillNames
# print(data)
```

```python
onehotMapping = []
for user in data.keys():
    userVector = []
    for skill in AllSkills:
        if(skill in data[user]):
            userVector.append(1)
        else:
            userVector.append(0)
    userVector.insert(0,user) #insert name at position 0
    onehotMapping.append(userVector)
```

```python
columns = []
columns.extend(AllSkills)
columns.insert(0,'Name')
columns
df = pd.DataFrame(onehotMapping,columns=columns)
df.head()
```

```python
#To create skill - domain mapping
SkillDomainMap = []
AllDomains = mapDict.keys()

for skill in AllSkills:
    skillvector = []
    for domain in AllDomains:
        if(skill in mapDict[domain]):
            skillvector.append(1)
        else:
            skillvector.append(0)
    skillvector.insert(0,skill)
    SkillDomainMap.append(skillvector)
```

```python
columns = []
columns.extend(AllDomains)
columns.insert(0,'Skillname')
df2 = pd.DataFrame(SkillDomainMap,columns=columns)
df2.head()
```
