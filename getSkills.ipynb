```python
newdf = pd.DataFrame()
```

```python
newdf['hash'] = df['hash']
newdf['Name'] = df['Name']
```

```python
skills = []
for line in df['Skills']:
    splitline = line.strip('][').split(',')
    skillset = []
    for word in splitline:
        word = word.replace("'","")
        skillset.append(word)
    skills.append(skillset)
```

```python
skills
```

```python
newdf['Skills'] = skills
```

```python
type(newdf['Skills'][0])
```

```python
newdf.to_csv('Dataset.csv',index=False)
```

```python
from collections import Counter
countSkills = Counter(s for sk in newdf['Skills'] for s in set(sk))
```

```python
uniqueSkills = list(countSkills.keys())
```

```python
print(f'There are {len(uniqueSkills)} unique skills')
```
