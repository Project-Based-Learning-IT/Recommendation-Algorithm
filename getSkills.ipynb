```python
import pandas as pd
newdf = pd.read_csv('Dataset.csv')
```

```python
newdf
```

```python
newdf['Skills'][0]
```

```python
for sk in newdf['Skills']:
    print(sk.strip('][').split(','))
```

```python
from collections import Counter
countSkills = Counter(s.replace("'","").strip() for sk in newdf['Skills'] for s in set(sk.strip('][').split(',')))
```

```python
type(countSkills)
```

```python
uniqueSkills = list(countSkills.keys())
```

```python
print(f'There are {len(uniqueSkills)} unique skills')
```

```python
singleItems = []
for item,freq in countSkills.items():
    if(freq == 1):
        singleItems.append(item)
```

```python
singleItems
```

```python
import json
```

```python
with open('skillCount.json','w') as f:
    json.dump(countSkills,f)
```

```python
df = pd.read_csv('Dataset.csv')
```

```python
df
```

```python
Totalskills = []
for line in df['Skills']:
    splitline = line.strip('][').split(',')
#     print(splitline)
    skillsPerUser = []
    for word in splitline:
        word = word.replace("'","").rstrip().lstrip().lower()
        if(word not in singleItems):
            skillsPerUser.append(word)
    Totalskills.append(skillsPerUser)
```

```python
df['Skills'] = Totalskills
```

```python
df.to_csv('SkippedSingleValuesDataset.csv',index=False)
```
