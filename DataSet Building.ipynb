```python
import json
import pandas as pd
```

```python
skills = json.load(open('skills.json'))
```

```python
SkillList = list(skills.keys())
```

```python
print('Total skills - {}'.format(len(SkillList)))
```

```python
df = pd.read_csv('SkippedSingleValuesDataset.csv')
```

```python
df
```

```python
SkillRepresentation = []
for skills_per_user in df['Skills'].to_list(): #will run 629 times
    userSkillArray = []
    for skill in SkillList: #will run 489 times
        if skill in skills_per_user:
            userSkillArray.append(1)
        else:
            userSkillArray.append(0)
    SkillRepresentation.append(userSkillArray)
```

```python
names = open('names.txt').read().strip('][').split(',')
```

```python
for i in range(len(SkillRepresentation)):
    SkillRepresentation[i].insert(0,names[i].replace("'",""))
```
